```python
import xml.etree.ElementTree as ET
import cv2
from matplotlib import pyplot as plt

# Function to parse the XML file and return the annotation details
def parse_annotation(annotation_path):
    tree = ET.parse(annotation_path)
    root = tree.getroot()
    objects = root.findall('object')
    boxes = []
    for object in objects:
        bbox = object.find('bndbox')
        boxes.append({
            'label': object.find('name').text,
            'xmin': int(bbox.find('xmin').text),
            'ymin': int(bbox.find('ymin').text),
            'xmax': int(bbox.find('xmax').text),
            'ymax': int(bbox.find('ymax').text)
        })
    return boxes

# Function to display an image with annotations
def display_image_with_annotations(image_path, annotation_path):
    # Read the image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Parse the annotation file
    annotations = parse_annotation(annotation_path)
    
    # Draw the boxes on the image
    for annotation in annotations:
        cv2.rectangle(image, 
                      (annotation['xmin'], annotation['ymin']), 
                      (annotation['xmax'], annotation['ymax']), 
                      (255, 0, 0), 2)
        cv2.putText(image, 
                    annotation['label'], 
                    (annotation['xmin'], annotation['ymin']-10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 
                    0.9, (255,0,0), 2)
    
    # Display the image
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.axis('off')
    plt.show()

# Display the first image along with its annotations
first_image_path = os.path.join(train_directory, train_images[0])
first_annotation_path = os.path.join(train_directory, train_annotations[0])
display_image_with_annotations(first_image_path, first_annotation_path)
```

```python
import numpy as np

# Set the dimensions for the patches
patch_size = (64, 64)  # Width x Height

# Function to generate patches from an image
def generate_patches(image_path, annotation_path, patch_size, stride=32):
    """
    This function generates patches from the image and labels them based on the annotation.
    :param image_path: Path to the image file
    :param annotation_path: Path to the corresponding annotation file
    :param patch_size: A tuple (width, height) of the patch size
    :param stride: The stride with which the window slides over the image
    :return: A list of image patches and their labels (defect or no defect)
    """
    # Read the image
    image = cv2.imread(image_path)
    
    # Parse the annotation file
    annotations = parse_annotation(annotation_path)
    
    # Initialize list of patches and labels
    patches = []
    labels = []
    
    # Slide a window over the image
    for y in range(0, image.shape[0] - patch_size[1] + 1, stride):
        for x in range(0, image.shape[1] - patch_size[0] + 1, stride):
            patch = image[y:y+patch_size[1], x:x+patch_size[0]]
            label = 0  # Default label is 'no defect'
            
            # Check if the patch intersects with any defect annotation
            for annotation in annotations:
                if (x < annotation['xmax'] and x + patch_size[0] > annotation['xmin'] and
                    y < annotation['ymax'] and y + patch_size[1] > annotation['ymin']):
                    label = 1  # Set label to 'defect'
                    break
                    
            patches.append(patch)
            labels.append(label)
            
    return np.array(patches), np.array(labels)

# Generate patches for the first image as an example
patches, labels = generate_patches(first_image_path, first_annotation_path, patch_size)

# Display some patches
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(patches[i])
    plt.title('Defect' if labels[i] == 1 else 'No Defect')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.utils import Sequence
from sklearn.utils import shuffle

class SteelDefectsDataGenerator(Sequence):
    'Generates data for Keras'
    def __init__(self, image_paths, annotation_paths, batch_size=32, patch_size=(64, 64), stride=32, shuffle=True):
        'Initialization'
        self.image_paths = image_paths
        self.annotation_paths = annotation_paths
        self.batch_size = batch_size
        self.patch_size = patch_size
        self.stride = stride
        self.shuffle = shuffle
        self.on_epoch_end()

    def __len__(self):
        'Denotes the number of batches per epoch'
        return int(np.floor(len(self.image_paths) / self.batch_size))

    def __getitem__(self, index):
        'Generate one batch of data'
        # Generate indexes of the batch
        indexes = self.indexes[index*self.batch_size:(index+1)*self.batch_size]

        # Find list of IDs
        image_paths_temp = [self.image_paths[k] for k in indexes]
        annotation_paths_temp = [self.annotation_paths[k] for k in indexes]

        # Generate data
        X, y = self.__data_generation(image_paths_temp, annotation_paths_temp)

        return X, y

    def on_epoch_end(self):
        'Updates indexes after each epoch'
        self.indexes = np.arange(len(self.image_paths))
        if self.shuffle == True:
            np.random.shuffle(self.indexes)

    def __data_generation(self, image_paths_temp, annotation_paths_temp):
        'Generates data containing batch_size samples' # X : (n_samples, *patch_size, n_channels)
        # Initialization
        X = []
        y = []

        # Generate data
        for i, (img_path, ann_path) in enumerate(zip(image_paths_temp, annotation_paths_temp)):
            # Generate patches and labels for the image
            patches, labels = generate_patches(img_path, ann_path, self.patch_size, self.stride)
            
            # Extend the existing list
            X.extend(patches)
            y.extend(labels)

        return np.array(X), np.array(y)

# Prepare the full paths for the training images and annotations
train_image_full_paths = [os.path.join(train_directory, img) for img in train_images]
train_annotation_full_paths = [os.path.join(train_directory, ann) for ann in train_annotations]

# Instantiate the data generator for the training data
batch_size = 16  # Adjust the batch size as needed
train_data_generator = SteelDefectsDataGenerator(train_image_full_paths, train_annotation_full_paths, batch_size=batch_size)

# Stride # The stride with which the window slides over the image
stride = 32
# Test the data generator with the first batch
X_sample, y_sample = train_data_generator.__getitem__(0)
print(f'Sample batch shape: {X_sample.shape}')
print(f'Sample labels: {y_sample[:10]}')  # Display the first 10 labels
```

```python
from tensorflow.keras.utils import Sequence
import numpy as np
import cv2

class SteelDefectsDataGenerator(Sequence):
    'Generates data for Keras'
    def __init__(self, image_paths, annotation_paths, batch_size=32, patch_size=(64, 64), stride=32, shuffle=True):
        'Initialization'
        self.image_paths = image_paths
        self.annotation_paths = annotation_paths
        self.batch_size = batch_size
        self.patch_size = patch_size
        self.stride = stride
        self.shuffle = shuffle
        self.on_epoch_end()

    def __len__(self):
        'Denotes the number of batches per epoch'
        return int(np.floor(len(self.image_paths) / self.batch_size))

    def __getitem__(self, index):
        'Generate one batch of data'
        # Generate indexes of the batch
        indexes = self.indexes[index*self.batch_size:(index+1)*self.batch_size]

        # Find list of IDs
        image_paths_temp = [self.image_paths[k] for k in indexes]
        annotation_paths_temp = [self.annotation_paths[k] for k in indexes]

        # Generate data
        X, y = self.__data_generation(image_paths_temp, annotation_paths_temp)

        return X, y

    def on_epoch_end(self):
        'Updates indexes after each epoch'
        self.indexes = np.arange(len(self.image_paths))
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __data_generation(self, image_paths_temp, annotation_paths_temp):
        'Generates data containing batch_size samples'
        # Initialize
        X, y = [], []
        # Generate data
        for img_path, ann_path in zip(image_paths_temp, annotation_paths_temp):
            # Generate patches and labels for the image
            patches, labels = generate_patches(img_path, ann_path, self.patch_size, self.stride)
            X.extend(patches)
            y.extend(labels)
        return np.array(X), np.array(y)
```

```python
# Instantiate the data generator
train_data_generator = SteelDefectsDataGenerator(train_image_full_paths, train_annotation_full_paths, batch_size, patch_size, stride)
```

```python
# Instantiate the data generator
test_data_generator = SteelDefectsDataGenerator(test_image_full_paths, test_annotation_full_paths, batch_size, patch_size, stride)
```

```python
# Instantiate the data generator
valid_data_generator = SteelDefectsDataGenerator(valid_image_full_paths, valid_annotation_full_paths, batch_size, patch_size, stride)
```

```python
# Create the model (example using a simple CNN)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')  # Binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_data_generator,
    epochs=10,
    validation_data=valid_data_generator
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
